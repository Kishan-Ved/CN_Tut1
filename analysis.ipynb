```python
import pandas as pd
```

```python
df = pd.read_csv("timings.csv")
```

```python
df
```

## Single-Process Server

```python
df1 = df[df["Type"]==1]
```

```python
df1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(df1["Clients"], df1["Time"], marker="o", linestyle="-", color="b", label="Single-Process")

# Labels and Title
plt.xlabel("Number of clients")
plt.ylabel("Execution Time (Latency)")
plt.title("Clients vs. Time for Single Process Server")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

## Multi-Process Server

```python
df2 = df[df["Type"]==2]

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(df2["Clients"], df2["Time"], marker="o", linestyle="-", color="b", label="Multi-Process")

# Labels and Title
plt.xlabel("Number of clients")
plt.ylabel("Execution Time (Latency)")
plt.title("Clients vs. Time for Multi Process Server")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

## Multi-Threaded Server

```python
df3 = df[df["Type"]==3]

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(df3["Clients"], df3["Time"], marker="o", linestyle="-", color="b", label="Multi-Threaded")

# Labels and Title
plt.xlabel("Number of clients")
plt.ylabel("Execution Time (Latency)")
plt.title("Clients vs. Time for Multi Threaded Server")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

## Multi-Process vs Multi-Threaded Server

```python
df3 = df[df["Type"]==3]

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(df2["Clients"], df2["Time"], marker="o", linestyle="-", color="b", label="Multi Process")
plt.plot(df3["Clients"], df3["Time"], marker="o", linestyle="-", color="orange", label="Multi Threaded")

# Labels and Title
plt.xlabel("Number of clients")
plt.ylabel("Execution Time (Latency)")
plt.title("Multi-Process vs Multi-Threaded Server")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
